##### Aplico un ciclo for para el calculo de estadistica descriptiva y otras metricas importantes. Estas metricas ayudan en el analisis de datos al proporcionar una vision rapida y resumida de la distribucion y dispersion de los datos, permitiendo identificar patrones, tendencias y posibles anomalias

```python
for i in archivo_data_stats.columns:
    print(i)
    print(archivo_data_stats[i].describe())
    print('-------------------------------------')
```

##### Se crea una funcion para obtener la media, la varianza y el rango.

```python
def describir_datos(data):
    for i in data.columns:
        if data[i].dtype == 'float64' or data[i].dtype == 'int64':
            print(i)
            print('Mediana:', data[i].median())
            print('Varianza:', data[i].var())
            print('Rango',data[i].max()-data[i].min())
            print('Moda',data[i].mode())
            print('---')

describir_datos(archivo_data_stats)
```
